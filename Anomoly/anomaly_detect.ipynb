```python
# Drop rows with any missing values
df_clean = df.dropna()
print('Shape after dropping missing:', df_clean.shape)
print('Any missing left?', df_clean.isnull().sum().sum())
```

## Feature Engineering
Now that the data is clean, the next step is to create new features that help the model detect anomalies. This may include:
- Extracting hour and day from timestamps
- Calculating bytes per second
- Encoding categorical variables (e.g., protocol)
- Handling ports and IPs if needed
- Any other domain-specific features

```python
# Feature Engineering Example
import numpy as np
import pandas as pd

# Use df_clean from previous step
df_feat = df_clean.copy()

# Extract hour and day of week from timestamp
df_feat['timestamp'] = pd.to_datetime(df_feat['timestamp'], errors='coerce')
df_feat['hour_of_day'] = df_feat['timestamp'].dt.hour
df_feat['day_of_week'] = df_feat['timestamp'].dt.dayofweek

# Calculate bytes per second (avoid division by zero)
df_feat['bytes_per_second'] = df_feat['bytes'] / df_feat['duration_sec'].replace(0, np.nan)

# Encode protocol as categorical (one-hot)
df_feat = pd.get_dummies(df_feat, columns=['protocol'], prefix='protocol')

# Show the first few rows of engineered features
df_feat.head()
```

## Label Preparation
Now, ensure your anomaly label column (e.g., `anomalystatus`) is present, correctly formatted (0 for normal, 1 for anomaly), and has no missing values. This is required for supervised model training.

```python
# Label Preparation: Check and process 'anomalystatus' column
print('Unique values in anomalystatus:', df_feat['anomalystatus'].unique())
print('Missing values in anomalystatus:', df_feat['anomalystatus'].isnull().sum())

# Drop rows with missing or invalid labels (not 0 or 1)
df_feat = df_feat[df_feat['anomalystatus'].isin([0, 1])]
df_feat = df_feat.dropna(subset=['anomalystatus'])
print('Shape after label cleaning:', df_feat.shape)
print('Label value counts:')
print(df_feat['anomalystatus'].value_counts())
```

## Model Training and Evaluation
Now split your data into train and test sets, train a supervised model (e.g., Random Forest), and evaluate its performance using accuracy, precision, recall, and f1-score.

```python
# Model Training and Evaluation Example: Random Forest
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Select features (drop columns not used for prediction)
feature_cols = [col for col in df_feat.columns if col not in ['timestamp','session_id','src_ip','dst_ip','anomalystatus','Phone Number A','Phone Number B']]
X = df_feat[feature_cols]
y = df_feat['anomalystatus']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Precision:', precision_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('F1 Score:', f1_score(y_test, y_pred))
print('\nClassification Report:\n', classification_report(y_test, y_pred))
```

```python
# Model Training and Evaluation: XGBoost and LightGBM
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# XGBoost
xgb = XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
print('--- XGBoost ---')
print('Accuracy:', accuracy_score(y_test, y_pred_xgb))
print('Precision:', precision_score(y_test, y_pred_xgb))
print('Recall:', recall_score(y_test, y_pred_xgb))
print('F1 Score:', f1_score(y_test, y_pred_xgb))
print('\nClassification Report:\n', classification_report(y_test, y_pred_xgb))

# LightGBM
lgbm = LGBMClassifier(n_estimators=100, random_state=42)
lgbm.fit(X_train, y_train)
y_pred_lgbm = lgbm.predict(X_test)
print('--- LightGBM ---')
print('Accuracy:', accuracy_score(y_test, y_pred_lgbm))
print('Precision:', precision_score(y_test, y_pred_lgbm))
print('Recall:', recall_score(y_test, y_pred_lgbm))
print('F1 Score:', f1_score(y_test, y_pred_lgbm))
print('\nClassification Report:\n', classification_report(y_test, y_pred_lgbm))
```

```python
# Advanced Models: CatBoost and Stacking Ensemble
from catboost import CatBoostClassifier
from sklearn.ensemble import StackingClassifier

# CatBoost (handles categorical features natively)
cat = CatBoostClassifier(iterations=100, verbose=0, random_state=42)
cat.fit(X_train, y_train)
y_pred_cat = cat.predict(X_test)
print('--- CatBoost ---')
print('Accuracy:', accuracy_score(y_test, y_pred_cat))
print('Precision:', precision_score(y_test, y_pred_cat))
print('Recall:', recall_score(y_test, y_pred_cat))
print('F1 Score:', f1_score(y_test, y_pred_cat))
print('\nClassification Report:\n', classification_report(y_test, y_pred_cat))

# Stacking Ensemble (combine RF, XGB, LGBM)
estimators = [
    ('rf', rf),
    ('xgb', xgb),
    ('lgbm', lgbm)
    ]
stack = StackingClassifier(estimators=estimators, final_estimator=CatBoostClassifier(iterations=50, verbose=0, random_state=42), n_jobs=-1)
stack.fit(X_train, y_train)
y_pred_stack = stack.predict(X_test)
print('--- Stacking Ensemble ---')
print('Accuracy:', accuracy_score(y_test, y_pred_stack))
print('Precision:', precision_score(y_test, y_pred_stack))
print('Recall:', recall_score(y_test, y_pred_stack))
print('F1 Score:', f1_score(y_test, y_pred_stack))
print('\nClassification Report:\n', classification_report(y_test, y_pred_stack))
```

```python
# Hyperparameter Tuning: Stacking Ensemble with GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
# Define parameter grid for the final estimator (CatBoost)
stack_param_grid = {
    'final_estimator__iterations': [50, 100],
    'final_estimator__depth': [4, 6],
    'final_estimator__learning_rate': [0.05, 0.1],
    'final_estimator__l2_leaf_reg': [1, 3]
}

stack_grid = GridSearchCV(
    stack,
    param_grid=stack_param_grid,
    scoring='f1',
    cv=3,
    n_jobs=-1,
    verbose=1
)
stack_grid.fit(X_train, y_train)

print('Best Stacking Ensemble parameters:', stack_grid.best_params_)

# Evaluate best stacking ensemble on test data
best_stack = stack_grid.best_estimator_
y_pred_best_stack = best_stack.predict(X_test)
print('Tuned Stacking Ensemble Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_best_stack))
print('Tuned Stacking Ensemble Classification Report:')
print(classification_report(y_test, y_pred_best_stack))
```

```python
# Hyperparameter Tuning: CatBoost with GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV

cat_param_grid = {
    'iterations': [100, 200],
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1],
    'l2_leaf_reg': [1, 3, 5]
}

cat_grid = GridSearchCV(CatBoostClassifier(verbose=0, random_state=42), cat_param_grid, scoring='f1', cv=3, n_jobs=-1, verbose=1)
cat_grid.fit(X_train, y_train)

print('Best CatBoost parameters:', cat_grid.best_params_)

# Evaluate best estimator on test data
best_cat = cat_grid.best_estimator_
y_pred_best_cat = best_cat.predict(X_test)
print('Tuned CatBoost Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_best_cat))
print('Tuned CatBoost Classification Report:')
print(classification_report(y_test, y_pred_best_cat))
```

## Advanced ML Techniques: Feature Selection, SMOTE, and Stacking Ensemble

We'll apply advanced techniques to further improve anomaly detection:

- **Feature Selection**: Use RFECV to select the most important features.
- **SMOTE**: Balance the classes using Synthetic Minority Over-sampling Technique.
- **Stacking Ensemble**: Combine tuned CatBoost, XGBoost, and Neural Network for robust predictions.

```python
# Advanced ML: Feature Selection, SMOTE, and Stacking Ensemble
from sklearn.feature_selection import RFECV
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import StackingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from scikeras.wrappers import KerasClassifier
from tensorflow import keras
from tensorflow.keras import layers

# 1. Feature Selection with RFECV (using CatBoost as estimator)
rfecv = RFECV(estimator=best_cat, step=1, cv=StratifiedKFold(3), scoring='f1', n_jobs=-1)
rfecv.fit(X_train, y_train)
X_train_fs = rfecv.transform(X_train)
X_test_fs = rfecv.transform(X_test)

print('Selected features:', list(X_train.columns[rfecv.support_]))

# 2. Balance classes with SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train_fs, y_train)

# 3. Stacking Ensemble: CatBoost, XGBoost, Neural Network (KerasClassifier)
def build_best_model():
    model = keras.Sequential([
        layers.Input(shape=(X_train_fs.shape[1],)),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(32, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

nn_stack = KerasClassifier(model=build_best_model, epochs=20, batch_size=32, verbose=0)
xgb_stack = XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss', random_state=42)
stack_ensemble = StackingClassifier(estimators=[
    ('cat', best_cat),
    ('xgb', xgb_stack),
    ('nn', nn_stack)
], final_estimator=best_cat, n_jobs=-1)

stack_ensemble.fit(X_train_bal, y_train_bal)
y_pred_stack = stack_ensemble.predict(X_test_fs)

print('--- Advanced Stacking Ensemble ---')
print('Accuracy:', accuracy_score(y_test, y_pred_stack))
print('Precision:', precision_score(y_test, y_pred_stack))
print('Recall:', recall_score(y_test, y_pred_stack))
print('F1 Score:', f1_score(y_test, y_pred_stack))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_stack))
print('Classification Report:')
print(classification_report(y_test, y_pred_stack))
```

## Advanced Neural Network: Deeper MLP with Regularization and Callbacks
Let's try a deeper neural network with more layers, batch normalization, higher dropout, and callbacks for early stopping and learning rate reduction. This can help boost performance on tabular data.

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Build a deeper neural network
mlp_deep = keras.Sequential([
    layers.Input(shape=(X_train_scaled.shape[1],)),
    layers.Dense(128, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.4),
    layers.Dense(64, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')
])
mlp_deep.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-5)

# Train the model
history_deep = mlp_deep.fit(
    X_train_scaled, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# Predict and evaluate
y_pred_mlp_deep = (mlp_deep.predict(X_test_scaled) > 0.5).astype(int).flatten()
print('--- Deep Neural Network (MLP, Advanced) ---')
print('Accuracy:', accuracy_score(y_test, y_pred_mlp_deep))
print('Precision:', precision_score(y_test, y_pred_mlp_deep))
print('Recall:', recall_score(y_test, y_pred_mlp_deep))
print('F1 Score:', f1_score(y_test, y_pred_mlp_deep))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_mlp_deep))
print('Classification Report:')
print(classification_report(y_test, y_pred_mlp_deep))
```

## Hyperparameter Tuning for Neural Network with KerasTuner
Let's use KerasTuner to automatically search for the best neural network architecture and training parameters for this anomaly detection task.

```python
# If not installed, uncomment the next line to install KerasTuner
# !pip install keras-tuner --quiet
import keras_tuner as kt

def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Input(shape=(X_train_scaled.shape[1],)))
    # Tune the number of layers and units
    for i in range(hp.Int('num_layers', 1, 3)):
        model.add(layers.Dense(hp.Int(f'units_{i}', min_value=32, max_value=256, step=32), activation='relu'))
        if hp.Boolean(f'batch_norm_{i}'):
            model.add(layers.BatchNormalization())
        model.add(layers.Dropout(hp.Float(f'dropout_{i}', 0.1, 0.5, step=0.1)))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=keras.optimizers.Adam(hp.Float('learning_rate', 1e-4, 1e-2, sampling='log')),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=1,
    directory='keras_tuner_dir',
    project_name='anomaly_nn_tuning'
)

early_stop = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-5)

tuner.search(
    X_train_scaled, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop, reduce_lr],
    verbose=2
)

# Get the best model and evaluate
best_model = tuner.get_best_models(num_models=1)[0]
y_pred_tuned = (best_model.predict(X_test_scaled) > 0.5).astype(int).flatten()
print('--- Tuned Neural Network (KerasTuner) ---')
print('Accuracy:', accuracy_score(y_test, y_pred_tuned))
print('Precision:', precision_score(y_test, y_pred_tuned))
print('Recall:', recall_score(y_test, y_pred_tuned))
print('F1 Score:', f1_score(y_test, y_pred_tuned))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_tuned))
print('Classification Report:')
print(classification_report(y_test, y_pred_tuned))
```

## Ensemble: Combine Neural Network, CatBoost, and TabNet
Ensembling predictions from your best models can further improve accuracy and robustness. We'll use a simple majority vote from the neural network, CatBoost, and TabNet.

```python
# Function to prepare model inputs from raw dataset rows
def prepare_model_input(row):
    """
    Prepare feature vector from a raw dataset row for the CatBoost model.
    Args:
        row (pd.Series): A row from the original dataset with columns like:
            timestamp, session_id, src_ip, src_port, dst_ip, dst_port, protocol,
            bytes, duration_sec, Phone Number A, Phone Number B, cell_tower_lat, cell_tower_lon, anomalystatus
    Returns:
        list: Feature vector in the order expected by the model (feature_cols)
    """
    # Extract basic features
    src_port = row['src_port']
    dst_port = row['dst_port']
    duration_sec = row['duration_sec']
    bytes_val = row['bytes']
    cell_tower_lat = row['cell_tower_lat']
    cell_tower_lon = row['cell_tower_lon']
    
    # Extract hour_of_day and day_of_week from timestamp
    timestamp = pd.to_datetime(row['timestamp'], errors='coerce')
    hour_of_day = timestamp.hour if pd.notna(timestamp) else 0
    day_of_week = timestamp.dayofweek if pd.notna(timestamp) else 0
    
    # Calculate bytes_per_second (avoid division by zero)
    bytes_per_second = bytes_val / duration_sec if duration_sec > 0 else 0
    
    # One-hot encode protocol
    protocol = row['protocol']
    protocol_onehot = [
        1 if protocol == 'HTTP' else 0,   # protocol_HTTP
        1 if protocol == 'HTTPS' else 0,  # protocol_HTTPS
        1 if protocol == 'SIP' else 0,    # protocol_SIP
        1 if protocol == 'TCP' else 0,    # protocol_TCP
        1 if protocol == 'UDP' else 0     # protocol_UDP
    ]
    
    # Combine all features in the correct order
    features = [
        src_port,
        dst_port,
        duration_sec,
        bytes_val,
        cell_tower_lat,
        cell_tower_lon,
        hour_of_day,
        day_of_week,
        bytes_per_second
    ] + protocol_onehot
    
    return features

# Example usage: Prepare input from the first row of the original dataset
# Assuming df is the original dataset loaded earlier
if 'df' in locals():
    sample_row = df.iloc[0]  # First row as example
    prepared_features = prepare_model_input(sample_row)
    print('Prepared features:', prepared_features)

    # Create DataFrame for prediction
    input_df = pd.DataFrame([prepared_features], columns=feature_cols)
    display(input_df)

    # Predict
    custom_input = input_df.values
    custom_pred = loaded_cat.predict(custom_input)
    print('Predicted class for sample row:', custom_pred[0])
else:
    print('Original dataset df not found. Load it first with: df = pd.read_csv("lemonrice_with_phone.csv", on_bad_lines="skip")')
```
